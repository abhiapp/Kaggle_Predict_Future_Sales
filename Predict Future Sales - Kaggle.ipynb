```python
train_data = pd.read_csv('sales_train_v2.csv', sep = ',')
train_data.head()
```

```python
# Convert date from string to date times
train_data['date'] = train_data['date'].apply(dateutil.parser.parse, dayfirst=True)
train_data.head()
```

```python
# Get the sum of the item_price per month
train_data.groupby(['shop_id','item_id','date'])['item_price'].sum()
```
